## Exercise 1: Contact Book 

1. **Define a function** called `add_contact` that:
   - Accepts three arguments: `contacts_dict`, `name`, and `phone_number`.
   - Adds the contact to `contacts_dict` with `name` as the key and `phone_number` as the value.

2. **Define another function** called `get_contact` that:
   - Takes two arguments: `contacts_dict` and `name`.
   - Returns the phone number of the specified contact if it exists; otherwise, returns `"Contact not found"`.

3. Test the functions by:
   - Creating an empty dictionary called `contacts`.
   - Adding three contacts to it using `add_contact`.
   - Trying to retrieve a contact's phone number using `get_contact`.

   **Example:**
   ```python
   contacts = {}
   add_contact(contacts, "Alice", "123-456-7890")
   add_contact(contacts, "Bob", "234-567-8901")
   print(get_contact(contacts, "Alice"))  # Should output: 123-456-7890
   print(get_contact(contacts, "Charlie"))  # Should output: Contact not found
   ```

## Exercise 2: Student Grades 

1. **Define a function** called `calculate_average` that:
   - Accepts a dictionary, `grades_dict`, where each key is a student's name and each value is their grade (integer).
   - Calculates and returns the average grade of all students.

2. **Define another function** called `find_top_student` that:
   - Takes the same `grades_dict` as input.
   - Finds and returns the name of the student with the highest grade.

3. Test the functions by:
   - Creating a dictionary called `grades` with at least five students and their grades.
   - Calling `calculate_average` to find the class average.
   - Calling `find_top_student` to find the student with the highest grade.

   **Example:**
   ```python
   grades = {
       "Alice": 88,
       "Bob": 95,
       "Charlie": 70,
       "Diana": 85,
       "Eve": 92
   }
   print(calculate_average(grades))  # Should output the average grade
   print(find_top_student(grades))  # Should output: Bob (or the student with the highest grade)
   ```
